# 04. Distplot (Numerical- Categorical)

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"])
sns.distplot(titanic[titanic["Survived"]==1]["Age"])
```

# 05.HeatMap (Categorical-Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic["Survived"])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic["Survived"]))
```

```python
titanic['Survived']=titanic['Survived'].map({0:0,1:1})
survival_rate=titanic.groupby('Pclass')['Survived'].mean()*100
survival_rate.plot(kind='bar')
#(titanic.groupby("Pclass").mean()["Survived"]*100).plot(kind='bar')
```

```python
titanic['Survived']=titanic['Survived'].map({0:0,1:1})
survival_rate=titanic.groupby('Sex')['Survived'].mean()*100
survival_rate.plot(kind='bar')
```

```python
titanic['Survived']=titanic['Survived'].map({0:0,1:1})
survival_rate=titanic.groupby('Embarked')['Survived'].mean()*100
survival_rate.plot(kind='bar')
```
