```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets/titanic.csv")
df.head()
```

```python
df.columns.unique()
```

```python
#Number denotes the number of family members travelling with that passenger
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot(kind='bar')
fig.set_title("Passengers travelling with")
fig.plot()
```

```python
df['number']
```

```python
#extracting numerical part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
df
```

```python
#extracting categorical part
```

```python
#Separates non numerical part with actual value and numerical denoted by nan
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num']=df['Cabin'].str.extract('(\d+)')
df['cabin_cat']=df["Cabin"].str[0]
df.head()
```

```python
df['cabin_cat'].unique()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
#extract the last bit of ticket as number
df['ticket_num']=df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

df['ticket_cat']=df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
df.head()
```
