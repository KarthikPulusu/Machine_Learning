# Working with Dates

```python
#Converting to datetime datatype
```

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

## Extract Year

```python
date['date_year']=date['date'].dt.year
date.sample(5)
```

## Extract Month

```python
date['date_month_no']=date['date'].dt.month
date.head()
```

```python
date['date_month_name']=date['date'].dt.month_name()
date.head()
```

## Extract Days

```python
date['date_day']=date['date'].dt.day
date.head(5)
```

```python
#day of week
date['date_dow']=date['date'].dt.dayofweek
date.head(5)
```

```python
#day od week name
date['date_dow_name']=date['date'].dt.day_name()
date.head(5)
```

```python
# date=date.drop(columns=['product_id','city_id','orders'])
```

```python
date["date_is_weekend"]=np.where(date['date_dow_name'].isin(['saturday','sunday']),1,0)
date.head()
```

```python
#is weekend
date['date_is_weekend']=np.where(date['date_dow_name'].isin (['Saturday','Sunday']),1,0)
date.head()
```

## Extract week of the year

```python
date.columns
```

```python
date['date_week']=date['date'].dt.isocalendar().week
date.head()
```

## Extracting quarter

```python
date['quarter']=date['date'].dt.quarter
date
```

Extract semester using quarter

```python
date['semester']=np.where(date['quarter'].isin([1,2]),1,2) #1 if itr is 1,2 else 2
date
```
